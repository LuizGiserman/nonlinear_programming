    Estudo da função

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
from pylab import meshgrid
import math
```

    Definindo a função
$$
sin(x_1² - x_2²) * cos(x_1² + x_2²)
$$

```python
def f(x1, x2):
    return (np.sin((x1**2)  - (x2**2)) * np.cos((x1**2) + (x2**2)))
```

    Plot abrangente da função

```python
x1 = np.linspace(-50, 50, 100)
x2 = np.linspace(-50, 50, 100)

X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
surf = ax.plot_surface(X1, X2, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('z');

fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

Olhando para uma seção local

```python
x1 = np.linspace(-2, 2, 20)
x2 = np.linspace(-2, 2, 20)

X1, X2 = np.meshgrid(x1, x2)
Z = f(X1, X2)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
surf = ax.plot_surface(X1, X2, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('z');

fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
